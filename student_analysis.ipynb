```python
import matplotlib.pyplot as plt

# Histogram
plt.hist(df['G3'])
plt.title("Final Grade Distribution")
plt.show()
```

```python
# Bar chart
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.title("Gender vs Performance")
plt.show()
```
